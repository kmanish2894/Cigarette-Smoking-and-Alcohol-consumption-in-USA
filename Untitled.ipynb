# 2. Analysis of variance

I have tried to find association between number of cigarettes smoked per month(response variable) and drank atleast one alcoholic drink in life(explanatory variable with 2 levels).Also the association between  drinking status( (explanatory variable with more than 2 levels) and number of cigarettes smoked per month(response variable) 

```python
import numpy
import pandas
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 
data = pandas.read_csv('F:/Data Science project/analysis/assignment/code book and data set/nesarc.csv', low_memory=False)
#setting variables you will be working with to numeric
data['S3AQ3B1'] = data['S3AQ3B1'].convert_objects(convert_numeric=True)
data['S3AQ3C1'] = data['S3AQ3C1'].convert_objects(convert_numeric=True)
data['CHECK321'] = data['CHECK321'].convert_objects(convert_numeric=True)
#subset data to young adults age 18 to 25 who have smoked in the past 12 months
sub1=data[(data['AGE']>=18) & (data['AGE']<=25) & (data['CHECK321']==1)]
#SETTING MISSING DATA
sub1['S3AQ3B1']=sub1['S3AQ3B1'].replace(9, numpy.nan)
sub1['S3AQ3C1']=sub1['S3AQ3C1'].replace(99, numpy.nan)
recode1 = {1: 30, 2: 22, 3: 14, 4: 5, 5: 2.5, 6: 1}
sub1['USFREQMO']= sub1['S3AQ3B1'].map(recode1)
#converting new variable USFREQMMO to numeric
sub1['USFREQMO']= sub1['USFREQMO'].convert_objects(convert_numeric=True)
# Creating a secondary variable multiplying the days smoked/month and the number of cig/per day
sub1['NUMCIGMO_EST']=sub1['USFREQMO'] * sub1['S3AQ3C1']
sub1['NUMCIGMO_EST']= sub1['NUMCIGMO_EST'].convert_objects(convert_numeric=True)
ct1 = sub1.groupby('NUMCIGMO_EST').size()
sub3 = sub1[['NUMCIGMO_EST', 'S2AQ1']].dropna()
model2 = smf.ols(formula='NUMCIGMO_EST ~ C(S2AQ1)', data=sub3).fit()
print (model2.summary())
print 'means for numcigmo_est by drank atleast one alcoholic drink in life'
m2= sub3.groupby('S2AQ1').mean()
print (m2)
print ('standard deviations for numcigmo_est by drank atleast one alcoholic drink in life')
sd2 = sub3.groupby('S2AQ1').std()
print (sd2)
model1 = smf.ols(formula='NUMCIGMO_EST ~ C(CONSUMER)', data=sub1)
results1 = model1.fit()
print (results1.summary())
sub2 = sub1[['NUMCIGMO_EST', 'CONSUMER']].dropna()
print ('means for numcigmo_est by major depression status')
m1= sub2.groupby('CONSUMER').mean()
print (m1)
print ('standard deviations for numcigmo_est by major depression status')
sd1 = sub2.groupby('CONSUMER').std()
print (sd1)
mc1=multi.MultiComparison(sub2['NUMCIGMO_EST'],sub2['CONSUMER'])
res1=mc1.tukeyhsd()
print (res1.summary())
```

explanatory variable:

NULL HYPOTHESIS:(Ho)

There is no association between no.of cigarettes smoked per month and drank atleast one alcoholic drink in life.

ALTERNATE HYPOTHESIS:(Ha)

There is an association between no.of cigarettes smoked per month and drank atleast one alcoholic drink in life. 

From the output we can see that the F-statistics for the above situation is 0.32 and the p-value is 0.57.

Since the p value 0.57>0.05 we shall accept null hypothesis

From the means we can say that young adult smokers who drank atleast one alcoholic drink in life smoke 321.2 cigarettes per month and young adult smokers who did not drank atleast one alcoholic drink in life smoke 304.7 cigarettes per month

Post hoc paired comparisons (i.e. more than two levels of a explanatory variable):

NULL HYPOTHESIS:(Ho)

There is no association between drinking status(i.e current drinker,ex-drinker and lifetime abstainer) and no.of cigarettes smoked per month 

ALTERNATE HYPOTHESIS:(Ha)

There is no association between drinking status(i.e current drinker,ex-drinker and lifetime abstainer) and no.of cigarettes smoked per month

The F-statistics and p-value for the above situation is 0.75 and 0.46 respectively.

Since the p value 0.46>0.05 we shall accept null hypothesis

From the means we can say that CONSUMER 3 which indicates lifetime abstainer smokes the lowest no.of cigarettes per month(304.7) and CONSUMER 2 which indicates ex-drinker smoke the highest no.of cigarettes per month (353.3).

Also it can be observed that the multiple comparison of means using Tukey HSD shows reject column as false for all comparisons i.e Current Drinker(1) Vs Ex-Drinker(2), Current Drinker(1) Vs Lifetime Abstainer(3) and Ex-Drinker(2) Vs Lifetime Abstainer(3) .

Finally it can be concluded that there is no association between no.of cigarettes smoked per month and drank atleast one alcoholic drink in life.Also there is no association between drinking status and no.of cigarettes smoked per month.

The Pearsons correlation coefficient for association between quantity and duration of cigarette smoking is 0.4033 which indicates that there exists a weak positive relationship between usual quantity when smoked cigarettes and duration of usual cigarette smoking.
